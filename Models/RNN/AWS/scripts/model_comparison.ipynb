```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
loss_df = pd.read_csv('/Users/ianshaw/Downloads/GitHub/deepcrypto/Models/RNN/model_compare/all_loss.csv')
```

```python
loss_df
```

```python
#Plot hidden layers vs. loss
loss_df = loss_df.loc[loss_df['Loss'] < 0.0005]
loss_df = loss_df.sort('Hidden_Size')
x_val = 'Hidden_Size'
plt.semilogy(loss_df[x_val].values, loss_df['Loss'].values)
plt.show()
```

```python
#Plot optimizer vs. loss
opt_df = loss_df.loc[loss_df['Opt'].isin(['SGDM'])]
x_val = 'Hidden_Size'
plt.semilogy(opt_df[x_val].values, opt_df['Loss'].values, label = 'SGDM')
opt_df = loss_df.loc[loss_df['Opt'].isin(['SGD'])]
plt.semilogy(opt_df[x_val].values, opt_df['Loss'].values, label = 'SGD')
opt_df = loss_df.loc[loss_df['Opt'].isin(['Adam'])]
plt.semilogy(opt_df[x_val].values, opt_df['Loss'].values, label = 'Adam')
plt.legend()
plt.show()
```

```python
#Plot Learning (learning rate * 10^3) vs. loss
x_val = 'Learning'
loss_df = loss_df.sort(x_val)
plt.semilogy(loss_df[x_val].values, loss_df['Loss'].values)
plt.show()
```
